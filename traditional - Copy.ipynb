```python
import pandas as pd
import numpy as np

data=pd.read_excel('D:\coding\PRML\hobby\cervical cancer\glcm_properties.xlsx')
data
```

```python
df2=pd.read_excel('Updated_db.xlsx')
#df2=df2.sort_values(by="CaseID")
df2
```

```python
df2.columns
```

```python
df2.dropna(subset=[" Lugol Image Name"],inplace=True)
df2
```

```python
df2=df2[[" Lugol Image Name","Iodine Uptake"]]
df2.sort_values(by=" Lugol Image Name")
df2
```

```python
import pandas as pd

# Assuming you have your two dataframes named df1 and df2

# Merge the dataframes based on the "Lugol Image Name" and "file name" columns
merged_df = pd.merge(df2, data, left_on=" Lugol Image Name", right_on="Image", how="inner")

# The how="inner" parameter ensures that only the rows with matching file names are included in the result.

# Now, merged_df contains the desired data with properties from both dataframes.
merged_df
```

```python
merged_df.drop(columns=" Lugol Image Name",inplace=True)
```

```python
merged_df.columns
```

```python
from sklearn.model_selection import train_test_split
X=merged_df.drop(columns=["Image","Iodine Uptake"],inplace=False)
Y=merged_df["Iodine Uptake"]
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
x_train
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
svm = SVC(kernel = 'linear', C = 1).fit(x_train, y_train) 
svm_predictions = svm.predict(x_test) 
  
# model accuracy for X_test   
accuracy = svm.score(x_test, y_test) 
# creating a confusion matrix 
cm = confusion_matrix(y_test, svm_predictions) 
print("Accuracy of SVM is:",accuracy)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=25)
model.fit(x_train,y_train)
preds=model.predict(x_test)
acc=accuracy_score(y_test,preds)
print("The Accuracy of randomforestclassifier is:", acc)

parameters = {
    "n_estimators":[5,10,50,100,250],
    "max_depth":[2,4,8,16,32,None]
}

from sklearn.model_selection import GridSearchCV
cv = GridSearchCV(model,parameters,cv=5)
cv.fit(x_train,y_train)
print("The best accuracy score obtained is:",cv.best_score_," at parameters ", cv.best_params_)
```

```python
import pickle
model=RandomForestClassifier(n_estimators=25)
model.fit(x_train,y_train)
preds=model.predict(x_test)
acc=accuracy_score(y_test,preds)
print("The Accuracy of randomforestclassifier is:", acc)
pickle.dump(model, open('lugols.pkl', 'wb'))
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(x_train,y_train)
preds=model.predict(x_test)
acc=accuracy_score(y_test,preds)
print("The Accuracy of Naive Bayes is:", acc)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
preds=model.predict(x_test)
acc=accuracy_score(y_test,preds)
print("The Accuracy of Logistic regression is:", acc)
```

```python

from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler,MinMaxScaler
import pickle

max_score=0
max_ind=-1
for i in range(1,148):
    pipe = make_pipeline(MinMaxScaler(), KNeighborsClassifier(n_neighbors=32))
    pipe.fit(x_train, y_train)  # apply scaling on training data
    score=pipe.score(x_test, y_test)
    if score>max_score:
        max_score=score
        max_ind=i
print(max_score,max_ind)
```

```python
from sklearn.linear_model import Perceptron

model=Perceptron()
model.fit(x_train,y_train)
preds=model.predict(x_test)
acc=accuracy_score(y_test,preds)

print("The accuracy of a Perceptron is:",acc)
```

```python
from sklearn.neural_network import MLPClassifier
clf=MLPClassifier(random_state=1,max_iter=300)
clf.fit(x_train,y_train)
preds=clf.predict(x_test)
acc=accuracy_score(y_test,preds)
print("The accuracy score obtained by MLP classifier is :",acc)
```
